## 3. Create Visualizations for Numerical Columns



### A Histogram (Distribution of Numerical Features)


```python
dataset.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()
```

## Box Plot (Detect Outliers)

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=dataset.select_dtypes(include=['number']))
plt.title("Box Plot of Numerical Features")
plt.show()
```

## 4. Analyze Churn Rates

```python

churn_counts = dataset['Churn'].value_counts()
print(churn_counts)

churn_percentage = dataset['Churn'].value_counts(normalize=True) * 100
print(churn_percentage)
```

## Visualizing Churn vs. Non-Churn Customers

```python
sns.countplot(x='Churn', data=dataset)
plt.figure(figsize=(4, 4))
plt.title('Churn vs. Non-Churn Customers')
plt.show()
```

## Box Plot of Monthly Charges vs. Churn

```python
plt.figure(figsize=(4, 4))
sns.boxplot(x='Churn', y='MonthlyCharges', data=dataset)
plt.title('Monthly Charges Distribution for Churned vs. Non-Churned Customers')
plt.show()
```

### Contract Type vs. Churn

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Contract', hue='Churn', data=dataset, palette='viridis')
plt.title('Contract Type vs. Churn')
plt.show()
```

# Tasks 4: Customer Segmentation Visualization

```python
bins = [0, 12, 36, dataset['tenure'].max()]
labels = ['0-12 months', '13-36 months', '37+ months']
dataset['Tenure Group'] = pd.cut(dataset['tenure'], bins=bins, labels=labels, right=True)

# Count customers in each tenure group
tenure_counts = dataset['Tenure Group'].value_counts()
print(tenure_counts)
```

```python

plt.figure(figsize=(4, 4))
plt.pie(tenure_counts, labels=tenure_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'], startangle=140)
plt.title('Customer Distribution by Tenure')
plt.show()
```

```python

tenure_avg_charges =dataset.groupby ('Tenure Group') ['MonthlyCharges'].mean().reset_index()
print(tenure_avg_charges)
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x = 'Tenure Group', y ='MonthlyCharges', data = tenure_avg_charges )
plt.title('Comparision of Average Monthly Charges Across Tenure Categories')
plt.show()
```
